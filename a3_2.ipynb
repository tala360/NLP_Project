```python
def generate_features(sentences,window=1):
    mydict={}
    for sentence in sentences:
        for i,token in enumerate(sentence):
            current=mydict.get(token,{})
            features=sentence[max(0,i-window):i]+sentence[i+1:i+window+1]
            for feature in features:
                current[feature]=current.get(feature,0)+1
            mydict[token]=current
    return mydict
```

The function `generate_features(sentences,window=1)` takes two arguments:<br/>
`sentences`: takes a list of sentences (where each sentence is a list of tokens)<br/>
`window`: a window size, the default window size is 1.<br/>
It outputs a dictionary of dictionaries. The key to the outermost dictionary is a word token. Its value is another dictionary, and its key is another word (co-occurence feature, window word) and its value is the number of times the word (key of the outermost dictionary) and the other word (the key to the internal dictionary) co-occur together, depending on the window size.<br/>
This helps with finding the juxtaposition between words by finding the words that occur within a specific window either side of the word.<br/>
If a window size is one for a specific word, we only look at one word before that specific word and one after it. If a window size is two, we look at two words before and after that word, etc...<br/>
